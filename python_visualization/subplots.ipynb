```python
# importing libraries 
import random 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from matplotlib.pyplot import figure
import pandas as pd 
```

```python
x= np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
# fig = plt.figure()
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
plt.show()
```

```python
# side by side plot 
fig,(axs1,axs2)= plt.subplots(1, 2,figsize=(10,5))
axs1.plot(x, y)

axs1.set_title('Axis [0, 0]')
fig.tight_layout(pad=6.0)

axs2.plot(x, y, 'tab:orange')
axs2.set_title('Axis [0, 1]')
```
